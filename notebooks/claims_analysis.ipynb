```python
# Part 3: Advanced Analysis with Joins

## Question 6: Claims with High Service Line Counts

### Merge Header+Line
merged_hl= df_header.merge(df_line, on="ProspectiveClaimId", how="left")

### Total Number of Service Lines per Claim
line_counts= merged_hl.groupby("ProspectiveClaimId").size()

### Filter for claims >= 5 lines
line_claims_ids = line_counts[line_counts >=5].index

### Display: ClaimId, Provider name, number of lines, and total charges
summary_all = (
    merged_hl[merged_hl["ProspectiveClaimId"].isin(line_claims_ids)]
    .groupby("ProspectiveClaimId")
    .agg(
        Provider=("BillingProviderNPI", "first"),
        NumLine=("LinePos", "count"),
        TotalCharges=("Charges", "sum")
    )
    .reset_index()
    .sort_values("NumLine", ascending=False)
)
display(summary_all)
```

```python
# Question 7: Dx-Procedure Combination
## Merge HEADER + LINE+ CODE
merged_hlc = merged_hl.merge(df_code, on="ProspectiveClaimId", how="left")
merged_hlc.head()


# Filter for CPT 99291 and find top dx codes
dx_for_99291 = (
    merged_hlc[merged_hlc["HCPCS"] == "99291"]["CodeValue"]
    .value_counts()
    .head(10)
)
display(dx_for_99291)
```

```python
# Question 8: Charges by Payer
## Merge HEADER + LINE
# Merge HEADER + LINE datasets
merged_hl = df_header.merge(df_line, on="ProspectiveClaimId", how="left")

merged_hl.head()

## Charges by Payer
payer_stats = merged_hl.groupby("PrimaryPayerName").agg(
    TotalCharges=("Charges", "sum"),
    AvgCharges=("Charges", "mean"),
    NumClaims=("ProspectiveClaimId", "nunique")
).sort_values("TotalCharges", ascending=False)
payer_stats.head(10)
```

```python
# Part 4: Creative Analysis

# Question 9: **Which diagnoses are associated with the highest total charges in the dataset?**

# Merge HEADER + LINE + CODE
merged_all = (
    df_header
    .merge(df_line, on="ProspectiveClaimId", how="left")
    .merge(df_code, on="ProspectiveClaimId", how="left")
)

merged_all.head()

# Total Charges by DX Code
dx_charges = (
    merged_all
    .groupby("CodeValue")
    .agg(TotalCharges=("Charges", "sum"),
    NumClaims= ("ProspectiveClaimId", "nunique"),
    NumLines= ("LinePos", "count")
    )
    .sort_values("TotalCharges", ascending=False)
    .head(10)
)
display(dx_charges)

## Bar Chart for DX Charges
ax = dx_charges["TotalCharges"].plot(
    kind="bar",
    figsize=(12,4),
    title="Top 10 Diagnoses by Total Charges",
    color="lightblue"
)

plt.ylabel("Total Charges ($)")
plt.xlabel("ICD-10 Diagnosis Code")

# Label exact values
for i, v in enumerate(dx_charges["TotalCharges"]):
    ax.text(i, v + 500, f"${v:,.0f}", ha='center', fontsize=9)

plt.show()

```
